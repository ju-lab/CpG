```python
import sys, os, gzip

for chr_num in range(1,24):
    print chr_num
    input_file = file("union_signature_chr%s.txt" %str(chr_num).replace("23","X"))
    output_file = file("final_signature_chr%s.vcf" %str(chr_num).replace("23","X"),"w")
    output_file.write('#CHROM\tPOS\tID\tREF\tALT\tAC\n')
    input_line = input_file.readline().strip()
    input_line = input_file.readline().strip()
    
    input_list = []
    ac_list = []

    while input_line:
        input_split = input_line.split('\t')
        if len(input_split[3]) > 1 or len(input_split[4]) > 1 :
            input_line = input_file.readline().strip()
        else:
            break
            
    prev_line = input_line
    input_line = input_file.readline().strip()
    
    while input_line:
        input_split = input_line.split('\t')
        if len(input_split[3]) > 1 or len(input_split[4]) > 1 :
            input_line = input_file.readline().strip()
            continue
        
        prev_split = prev_line.split('\t')
        prev_pos = int(prev_split[1])
        input_pos = int(input_split[1])
        
        if input_pos == prev_pos:
            if prev_split[0:5] not in input_list:
                input_list.append(prev_split[0:5])
                ac_list.append(int(prev_split[5]))
            else:
                ac_list[input_list.index(prev_split[0:5])] += int(prev_split[5])

            prev_line = input_line
            input_line = input_file.readline().strip()        
            if input_line == '':
                if prev_split[0:5] not in input_list:
                    input_list.append(prev_split[0:5])
                    ac_list.append(int(prev_split[5]))
                else:
                    ac_list[input_list.index(prev_split[0:5])] += int(prev_split[5])
                    
                for i in range(0,len(input_list)):
                    if ac_list[i] < 44:
                        output_file.write('\t'.join(input_list[i])+'\t'+str(ac_list[i])+'\n')
                break
                
        else:
            if len(input_list) > 0:
                if prev_split[0:5] not in input_list:
                    input_list.append(prev_split[0:5])
                    ac_list.append(int(prev_split[5]))
                else:
                    ac_list[input_list.index(prev_split[0:5])] += int(prev_split[5])
                    
                for i in range(0,len(input_list)):
                    if ac_list[i] < 44:
                        output_file.write('\t'.join(input_list[i])+'\t'+str(ac_list[i])+'\n')                        
                    
                input_list = []
                ac_list = []

                prev_line = input_line
                input_line = input_file.readline().strip()                
                
            else:
                output_file.write(prev_line+'\n')
                prev_line = input_line
                input_line = input_file.readline().strip()                
                if input_line == '':
                    output_file.write(prev_line+'\n')
                    break                    
    
    input_file.close()
    output_file.close()
print 'THE END'
```
